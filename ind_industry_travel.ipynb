```python
#import libraries and raw travel industry data for india
import pandas as pd
data = pd.read_excel('ind_travel.xlsx')
data.to_csv()
data.head()
```

```python
#drop the first unused row and reset the index
data = data.drop(data.index[[0]])
data = data.reset_index()
data = data.drop('index',1)
data.rename(columns={'Date':'date'},inplace=True)
data.head()
```

```python
#update the dataset till 26th May 2020
new_row = {'date':'2020-05-15','thomas_close':24.30,'bls_close':29.45, 'transcorp_close':6.66,
           'cox_close':1.10, 'intl_close':38.55, 'crown_close':17.40}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-18','thomas_close':23.10,'bls_close':27.70, 'transcorp_close':6.97,
           'cox_close':1.05, 'intl_close':40.25, 'crown_close':17.65}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-19','thomas_close':22.60,'bls_close':27.35, 'transcorp_close':6.70,
           'cox_close':1, 'intl_close':41.55, 'crown_close':17}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-20','thomas_close':22.65,'bls_close':27.60, 'transcorp_close':6.94,
           'cox_close':0.95, 'intl_close':40.05, 'crown_close':17.45}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-21','thomas_close':23.55,'bls_close':28.75, 'transcorp_close':6.94,
           'cox_close':0.94, 'intl_close':41.05, 'crown_close':17.50}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-22','thomas_close':24.6,'bls_close':29.10, 'transcorp_close':6.89,
           'cox_close':0.90, 'intl_close':39.95, 'crown_close':17.45}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-26','thomas_close':25.8,'bls_close':31.25, 'transcorp_close':6.84,
           'cox_close':0.86, 'intl_close':38.70, 'crown_close':17.45}
data = data.append(new_row,ignore_index=True)
```

```python
#change the date to required format step 1
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%d-%m-%Y')
data['date'].unique()
```

```python
#change the date to required format step 2
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%Y-%m-%d')
data['date'].unique()
```

```python
data.columns
```

```python
#calculating the index
data['initial_calc_index'] = (((data['thomas_close'])*(data['thomas_volume'])+(data['bls_close'])*(data['bls_volume'])
                       +(data['transcorp_close'])*(data['transcorp_volume'])+(data['cox_close'])*(data['cox_volume'])
                       +(data['crown_close'])*(data['crown_volume'])+(data['intl_close'])*(data['intl_volume']))
                      /(data['thomas_volume']+data['bls_volume']+data['transcorp_volume']
                        +data['cox_volume']+data['crown_volume']+data['intl_volume']))
data['initial_calc_index'] = data['initial_calc_index'].round(2)
data['calc_index'] = (((data['thomas_close'])*(378.275)+(data['bls_close'])*(102.45)
                       +(data['transcorp_close'])*(31.783)+(data['cox_close'])*(72.855)
                       +(data['intl_close'])*(7.995))
                      /(378.275+102.45+31.783+72.855+7.995))
data['calc_index'] = data['calc_index'].round(2)
data.head()
```

```python
#save the file in csv format
data.to_csv('ind_travel_final.csv')
```
