```python
#import libraries and raw food industry data for india
import pandas as pd
data = pd.read_excel('ind_food.xlsx')
data.to_csv()
data.head()
```

```python
#drop the first unused row and reset the index
data = data.drop(data.index[[0]])
data = data.reset_index()
data = data.drop('index',1)
data.rename(columns={'Date':'date'},inplace=True)
data.head()
```

```python
#update the dataset till 26th May 2020
new_row = {'date':'2020-05-15','krbl_close':189.45,'ltfoods_close':21.20,
           'chaman_close':38.75, 'himalaya_close':7.45, 'kohi_close':8.10, 'klrf_close':28.8}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-18','krbl_close':183.55,'ltfoods_close':21.15,
           'chaman_close':38.80, 'himalaya_close':7.79, 'kohi_close':7.75, 'klrf_close':28.8}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-19','krbl_close':180.50,'ltfoods_close':20.55,
           'chaman_close':37.75, 'himalaya_close':8.17, 'kohi_close':7.90, 'klrf_close':29.5}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-20','krbl_close':181.95,'ltfoods_close':20.50,
           'chaman_close':38.50, 'himalaya_close':8.57, 'kohi_close':8.10, 'klrf_close':28}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-21','krbl_close':179.75,'ltfoods_close':20.80,
           'chaman_close':37.85, 'himalaya_close':8.99, 'kohi_close':7.98, 'klrf_close':29.45}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-22','krbl_close':178.80,'ltfoods_close':21.00,
           'chaman_close':38.85, 'himalaya_close':8.55, 'kohi_close':7.88, 'klrf_close':29.90}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-26','krbl_close':179.60,'ltfoods_close':22.00,
           'chaman_close':38.15, 'himalaya_close':8.13, 'kohi_close':8, 'klrf_close':29.90}
data = data.append(new_row,ignore_index=True)
```

```python
#change the date to required format step 1
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%d-%m-%Y')
data['date'].unique()
```

```python
#change the date to required format step 2
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%Y-%m-%d')
data['date'].unique()
```

```python
data.columns
```

```python
#calculating the index
data['initial_calc_index'] = (((data['krbl_close'])*(data['krbl_volume'])+(data['ltfoods_close'])*(data['ltfoods_volume'])
                       +(data['chaman_close'])*(data['chaman_volume'])+(data['himalaya_close'])*(data['himalaya_volume'])
                       +(data['kohi_close'])*(data['kohi_volume'])+(data['klrf_close'])*(data['klrf_volume']))
                      /(data['krbl_volume']+data['ltfoods_volume']+data['chaman_volume']
                        +data['himalaya_volume']+data['kohi_volume']+data['klrf_volume']))
data['initial_calc_index'] = data['initial_calc_index'].round(2)
data['calc_index'] = (((data['krbl_close'])*(235.39)+(data['ltfoods_close'])*(319.845)
                       +(data['chaman_close'])*(51.734)+(data['himalaya_close'])*(57.843)
                       +(data['kohi_close'])*(35.242)+(data['klrf_close'])*(5.541))
                      /(235.39+319.845+51.734+57.843+35.242+5.541))
data['calc_index'] = data['calc_index'].round(2)
data.head()
```

```python
#save the file in csv format
data.to_csv('ind_food_final.csv')
```
