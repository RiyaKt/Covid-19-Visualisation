```python
#import libraries and raw entertainment industry data for india
import pandas as pd
data = pd.read_csv('Nifty Media Historical Data.csv')
#data.to_csv()
data.head()
```

```python
#rename the coulmns for uniformity across datasets
data.rename(columns={'Date':'date'},inplace=True)
data.rename(columns={'Price':'price'},inplace=True)
data.head()
```

```python
data.dtypes
```

```python
#change the date to required format
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%Y-%m-%d')
data['date'].unique()
```

```python
#remove the commas from price column and remove the unwanted columns
import numpy as np
data['price'] = data['price'].str.replace(',','').astype(np.float64)
data = data.drop('Open',1)
data = data.drop('High',1)
data = data.drop('Low',1)
data = data.drop('Vol.',1)
data = data.drop('Change %',1)
data.head()
```

```python
#save the file in csv format
data.to_csv('ind_entertainment.csv')
```
