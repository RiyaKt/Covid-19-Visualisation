```python
#import the raw data file
import pandas as pd
data = pd.read_csv('covid_19_india.csv')
data.head()
```

```python
#remove the columns that are not needed and print the list of unique dates
data = data.drop('Sno',1)
data = data.drop('Time',1)
data = data.drop('ConfirmedIndianNational',1)
data = data.drop('ConfirmedForeignNational',1)
#change the names of the columns to maintain unifromity across datasets and avoid confusion
data.rename(columns={'State/UnionTerritory':'state'},inplace=True)
data.rename(columns={'Date':'date'},inplace=True)
data.rename(columns={'Cured':'cum_recovered'},inplace=True)
data.rename(columns={'Deaths':'cum_deaths'},inplace=True)
data.rename(columns={'Confirmed':'cum_total'},inplace=True)
data.head()
data['date'].unique()
```

```python
#conversion of dates to the common date format step 1
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%d-%m-%Y')
data.head()
data['date'].unique()
```

```python
#conversion of dates to the common date format step 2
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%Y-%m-%d')
data['date'].unique()
```

```python
#find the state names mentioned as they should match the topographic json file
data['state'].unique()
```

```python
#make required changes in the spellings of states' names
data['state'] = data['state'].replace('Telengana','Andhra Pradesh')
data['state'] = data['state'].replace('Ladakh','Jammu and Kashmir')
data['state'] = data['state'].replace('Jharkhand#','Jharkhand')
data['state'] = data['state'].replace('Nagaland#','Nagaland')
data['state'] = data['state'].replace('Dadar Nagar Haveli','Dadra and Nagar Haveli')
data['state'].unique()
```

```python
#check how many unassigned cases are present in the dataset
data['state'].str.contains('Unassigned').value_counts()
```

```python
data['state'].str.contains('Cases being reassigned to states').value_counts()
```

```python
#group by the state names to avoid duplicates
df = data.groupby(['state','date']).agg('sum').reset_index()
```

```python
#import the statecode file
code = pd.read_excel('ind_statecode.xlsx')
code.to_csv()
code.rename(columns={'STATE CODE':'statecode','STATE NAME':'state'},inplace=True)
code
```

```python
#count number of rows for verification after merge
#merge the cases file with the statecode file
nfrows = df.count
print(nfrows)
final = pd.merge(df,code)
```

```python
#check the number of rows
#we find that the unassigned cases are removed after this step
nfrows = final.count
print(nfrows)
final.head()
```

```python
#save the file in csv format
final.to_csv('ind_state_case_final.csv')
```
