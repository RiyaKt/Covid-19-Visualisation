```python
#import the raw data files of all the industries
import pandas as pd
data1 = pd.read_csv('DAX Food & Beverage Historical Data.csv')
data2 = pd.read_csv('DAX Media Historical Data.csv')
data3 = pd.read_csv('DAX Retail Historical Data.csv')
data4 = pd.read_csv('LHAG Historical Data.csv')
data5 = pd.read_csv('TUI1d Historical Data.csv')
data5.head()
```

```python
#copy the relevant data into a new dataframe
data = data1
data.rename(columns={'Date':'date'},inplace=True)
data = data.drop('Price',axis=1)
data = data.drop('Open',axis=1)
data = data.drop('High',axis=1)
data = data.drop('Low',axis=1)
data = data.drop('Vol.',axis=1)
data = data.drop('Change %',axis=1)
#data['date'] = data1['Date']
data['food'] = data1['Price']
data['entertainment'] = data2['Price']
data['retail'] = data3['Price']
data['airline'] = data4['Price']
data['travel'] = data5['Price']
data.head()
```

```python
#change the format of the date to required fromat step 1
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%Y-%m-%d')
data['date'].unique()
```

```python
#change the format of the date to required fromat step 2
data['date'] = pd.to_datetime(data['date'])
data['date'].unique()
```

```python
#save the file in csv format
data.to_csv('ger_industry.csv')
```
