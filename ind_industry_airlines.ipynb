```python
#import libraries and raw airlines industry data for india
import pandas as pd
data = pd.read_excel('ind_airlines.xlsx')
data.to_csv()
data.head()
```

```python
#drop the first unused row and reset the index
data = data.drop(data.index[[0]])
data = data.reset_index()
data = data.drop('index',1)
data.rename(columns={'Date':'date'},inplace=True)
data.head()
```

```python
data.dtypes
```

```python
#update the dataset till 26th May 2020
new_row = {'date':'2020-05-15','spice_close':45.70,'jet_close':20.85,'interglobe_close':983.75, 'global_close':47.55}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-18','spice_close':43.45,'jet_close':19.90,'interglobe_close':912.20, 'global_close':50.05}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-19','spice_close':41.30,'jet_close':19.0,'interglobe_close':920.35, 'global_close':47.30}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-20','spice_close':40.95,'jet_close':19.35,'interglobe_close':911.9, 'global_close':46.20}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-21','spice_close':42.95,'jet_close':20.30,'interglobe_close':978.25, 'global_close':55.40}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-22','spice_close':45.05,'jet_close':21.30,'interglobe_close':974.10, 'global_close':55.85}
data = data.append(new_row,ignore_index=True)
new_row = {'date':'2020-05-26','spice_close':44.05,'jet_close':21.80,'interglobe_close':941.0, 'global_close':50.60}
data = data.append(new_row,ignore_index=True)
```

```python
data.fillna(0)
```

```python
#change the date to required format step 1
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%d-%m-%Y')
data['date'].unique()
```

```python
#change the date to required format step 2
data['date'] = data['date'].astype('datetime64[ns]')
data['date'] = data['date'].dt.strftime('%Y-%m-%d')
data['date'].unique()
```

```python
data.columns
```

```python
#calculating the index
data['initial_calc_index'] = (((data['spice_close'])*(data['spice_volume'])+(data['jet_close'])*(data['jet_volume'])
                       +(data['interglobe_close'])*(data['interglobe_volume'])+
                       (data['global_close'])*(data['global_volume']))/(data['spice_volume']+data['jet_volume']
                                                                        +data['interglobe_volume']+data['global_volume']))
data['initial_calc_index'] = data['initial_calc_index'].round(2)
data['calc_index'] = (((data['spice_close'])*(600.076)+(data['jet_close'])*(113.597)+(data['interglobe_close'])*(384.796)+(data['global_close'])*(14))/(600.076+113.597+384.796+14))
data['calc_index'] = data['calc_index'].round(2)
data.head()
```

```python
#save the file in csv format
data.to_csv('ind_airlines_final.csv')
```
