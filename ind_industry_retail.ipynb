```python
#index calculation
data['initial_calc_index'] = (((data['avenue_close'])*(data['avenue_volume'])+(data['aditya_close'])*(data['aditya_volume'])
                       +(data['vmart_close'])*(data['vmart_volume'])+(data['shoppers_close'])*(data['shoppers_volume'])
                       +(data['future_close'])*(data['future_volume'])+(data['afl_close'])*(data['afl_volume'])
                       +(data['trent_close'])*(data['trent_volume']))
                      /(data['avenue_volume']+data['aditya_volume']+data['vmart_volume']
                        +data['shoppers_volume']+data['future_volume']+data['afl_volume']+data['trent_volume']))
data['initial_calc_index'] = data['initial_calc_index'].round(2)
data['calc_index'] = (((data['avenue_close'])*(647.775)+(data['aditya_close'])*(773.948)
                       +(data['vmart_close'])*(18.156)+(data['shoppers_close'])*(87.99)
                       +(data['future_close'])*(527.398)+(data['afl_close'])*(78.655)
                       +(data['trent_close'])*(355.487))
                      /(647.775+773.948+18.156+87.99+527.398+78.655+355.487))
data['calc_index'] = data['calc_index'].round(2)
data.head()
```

```python
#save the file in csv format
data.to_csv('ind_retail_final.csv')
```
