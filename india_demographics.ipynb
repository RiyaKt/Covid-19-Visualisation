```python
#import libraries and raw demographics file
#since the raw file is in excel format, convert it into csv format
import pandas as pd
demo = pd.read_excel('india_demographics.xls')
demo.to_csv()
demo.head(10)
```

```python
#drop the extra rows
df = demo.drop(demo.index[[0,1,2,3,4]])
df.head(10)
```

```python
#drop the unwanted columns
df = df.drop('Unnamed: 0',1)
df = df.drop('Unnamed: 8',1)
df = df.drop('Unnamed: 9',1)
df = df.drop('Unnamed: 10',1)
df = df.drop('Unnamed: 11',1)
df = df.drop('Unnamed: 12',1)
df = df.drop('Unnamed: 13',1)
#rename the columns
df.rename(columns = {'Unnamed: 1':'statecode'},inplace=True)
df.rename(columns = {'Unnamed: 2':'towncode'},inplace=True)
df.rename(columns = {'Unnamed: 3':'areaname'},inplace=True)
df.rename(columns = {'C-14 CITY:  POPULATION IN FIVE YEAR AGE-GROUP BY RESIDENCE AND SEX - 2011 ':'agegroup'},inplace=True)
df.rename(columns = {'Unnamed: 5':'totalpopulation'},inplace=True)
df.rename(columns = {'Unnamed: 6':'male'},inplace=True)
df.rename(columns = {'Unnamed: 7':'female'},inplace=True)
df.head()
```

```python
df.head()
df['statecode'].unique()
```

```python
#reset the index step 1
df = df.reset_index()
df.head()
```

```python
#reset the index step 2
df = df.drop('index',1)
df.head()
```

```python
#remove the extra details in parenthesis from the town names
df['areaname'] = df['areaname'].str.replace(r"\(.*\)","")
df.head()
```

```python
#import the statecode file
code = pd.read_excel('ind_statecode.xlsx')
code.rename(columns={'STATE CODE':'statecode','STATE NAME':'state'},inplace=True)
code.to_csv()
code
```

```python
#convert statecode column in raw file to numeric for correct merging with the statecode file
df = df.convert_objects(convert_numeric=True)
df.dtypes
```

```python
#count the number of rows for verification later and merge the datasets
nfrows = df.count
print(nfrows)
final = pd.merge(df,code)
```

```python
#verify the number of rows after merging
nfrows = final.count
print(nfrows)
```

```python
#now the file contains statename as a column
final.head()
```

```python
#save the file in csv format
final.to_csv('ind_demo_final.csv')
```

```python
#check for missing states in the file
final['state'].unique()
#we find out 5 missing states/union territories
#Arunachal Pradesh, Daman and Diu, Dadra and Nagar Haveli, Goa, Lakshadweep
```

```python
#in further steps, we will convert the data into state-wise attributes required for the data model
data = pd.read_csv('ind_demo_final.csv')
data = data.drop('Unnamed: 0',1)
data = data.drop('areaname',1)
data = data.drop('towncode',1)
data = data.drop('statecode',1)
data.head()
```

```python
#use group by to convert data into state-wise format and drop all ages since it is the sum of all age group data
data = data.groupby(['state','agegroup']).agg('sum').reset_index()
#set the index as age-group to remove particular age groups when sorting into categories as and when required
data = data.set_index('agegroup')
data = data.drop('All ages',axis=0)
data.head(20)
```

```python
#data1 stores the total,male and female population of a particular state
data1 = data.groupby(['state']).agg('sum').reset_index()
data1.head(30)
```

```python
#data2 stores the state-wise oldage population numbers
data2 = data.drop(['Age not stated','0-4','5-9','10-14','15-19','20-24','25-29','30-34','35-39','40-44','45-49','50-54','55-59'],axis=0)
data2 = data2.groupby(['state']).agg('sum').reset_index()
data1['oldage'] = data2['totalpopulation']
data1.head()
```

```python
#data3 stores the state-wise adults population numbers
data3 = data.drop(['Age not stated','0-4','5-9','10-14','60-64','65-69','70-74','75-79','80+'],axis=0)
data3 = data3.groupby(['state']).agg('sum').reset_index()
data1['adults'] = data3['totalpopulation']
data1.head()
```

```python
#data4 stores the state-wise children population numbers
data4 = data.drop(['Age not stated','15-19','20-24','25-29','30-34','35-39','40-44','45-49','50-54',
                   '55-59','60-64','65-69','70-74','75-79','80+'],axis=0)
data4 = data4.groupby(['state']).agg('sum').reset_index()
data1['children'] = data4['totalpopulation']
data1.head()
```

```python
#append the rows for the 5 missing states
#source: check 5th entry in section 1 of documentation
new_row1 = {'state' : 'Arunachal Pradesh','totalpopulation' : 1383727, 'male' : 713912, 'female' : 669815,
            'oldage' : 63639, 'adults' : 825218, 'children' : 493361}
data1 = data1.append(new_row1,ignore_index=True)
new_row2 = {'state' : 'Daman and Diu','totalpopulation' : 243247 , 'male' : 150301, 'female' : 92946,
            'oldage' : 11361, 'adults' : 176545, 'children' : 54985}
data1 = data1.append(new_row2,ignore_index=True)
new_row3 = {'state' : 'Dadra and Nagar Haveli','totalpopulation' : 343709 , 'male' : 193760, 'female' : 149949,
            'oldage' : 13892, 'adults' : 221632, 'children' : 107813}
data1 = data1.append(new_row3,ignore_index=True)
new_row4 = {'state' : 'Goa','totalpopulation' : 1458545 , 'male' : 739140, 'female' : 719405,
            'oldage' : 163495, 'adults' : 974796, 'children' : 318160}
data1 = data1.append(new_row4,ignore_index=True)
new_row5 = {'state' : 'Lakshadweep','totalpopulation' : 64473 , 'male' : 33123, 'female' : 31350,
            'oldage' : 5270, 'adults' : 42631, 'children' : 16457}
data1 = data1.append(new_row5,ignore_index=True)
data1
```

```python
#percentage calculation
data1['percent_male'] = 100*data1['male']/data1['totalpopulation']
data1['percent_female'] = 100*data1['female']/data1['totalpopulation']
data1['percent_oldage'] = 100*data1['oldage']/(data1['oldage']+data1['adults']+data1['children'])
data1['percent_adults'] = 100*data1['adults']/(data1['oldage']+data1['adults']+data1['children'])
data1['percent_children'] = 100*data1['children']/(data1['oldage']+data1['adults']+data1['children'])
#rounding-off percentage to 2 decimal places
data1['percent_male'] = data1['percent_male'].round(2)
data1['percent_female'] = data1['percent_female'].round(2)
data1['percent_oldage'] = data1['percent_oldage'].round(2)
data1['percent_adults'] = data1['percent_adults'].round(2)
data1['percent_children'] = data1['percent_children'].round(2)
data1
data1
```
