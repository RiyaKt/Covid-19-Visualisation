```python
#the nest part involves the conversion of new cases to cumulative cases. This required the insertion of zero rows for all the
#dates, states and counties. To save computational time and storage, we have two separate files for states and counties
#as doing it in three nested for loops would require more time.
df = pd.read_csv('ger_case_final.csv')
```

```python
#first we prepare the state-wise cases file and hence we drop the county column
df = df.drop('Unnamed: 0',axis=1)
df = df.drop('code',axis=1)
df = df.drop('county',axis=1)
df.head()
```

```python
#group the data by state and set state and dates list
df = df.groupby(['state','date']).agg(sum).reset_index()
states = df['state'].unique()
dates = pd.date_range(df['date'].min(),df['date'].max())
#count number of rows for verification after the insertion of zero rows
nrows = df.count()
print(nrows)
df.head()
```

```python
#convert the data column from 'object' to 'date' format required for comparison operator used in the loop in next step
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
#loop to insert the zero-rows
for i in states:
    for k in dates:
        if(((df['state']==i) & (df['date']==k)).any()==False):
                new_row = {'state':i,'date':k,'new_total':0,'new_deaths':0,'new_recovered':0}
                df = df.append(new_row,ignore_index=True)
```

```python
#count the number of rows for verification
nrows = df.count()
print(nrows)
```

```python
#calculation for active cases and cumulative total cases, deaths, recovered, active cases
df['cum_total'] = df.groupby(by=['state'])['new_total'].transform(lambda x: x.cumsum())
df['cum_deaths'] = df.groupby(by=['state'])['new_deaths'].transform(lambda x: x.cumsum())
df['cum_recovered'] = df.groupby(by=['state'])['new_recovered'].transform(lambda x: x.cumsum())
df['cum_active']=df['cum_total']-df['cum_deaths']-df['cum_recovered']
df['new_active']=df['new_total']-df['new_deaths']-df['new_recovered']
pd.set_option('display.max_rows', None)
df.head()
```

```python
#save the file in csv format
df.to_csv('ger_covid_state.csv')
```

```python
#now we will do the same steps to obtain the county-wise cases file
#drop the state column in this case
df = pd.read_csv('ger_case_final.csv')
df = df.drop('Unnamed: 0',axis=1)
df = df.drop('code',axis=1)
df = df.drop('state',axis=1)
df.head()
```

```python
#prepare lists for counties and dates
#check the number of rows for later verification
counties = df['county'].unique()
dates = pd.date_range(df['date'].min(),df['date'].max())
nrows = df.count()
print(nrows)
df.head()
```

```python
#conver date column from object to date datatype for successful comparison in the loop
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
#insert zero-rows using two nested for loops
for j in counties:
    for k in dates:
        if(((df['county']==j) & (df['date']==k)).any()==False):
                new_row = {'county':j,'date':k,'new_total':0,'new_deaths':0,'new_recovered':0}
                df = df.append(new_row,ignore_index=True)
```

```python
#count the number of rows for verification
nrows = df.count()
print(nrows)
```

```python
#calculation for active cases and cumulative total cases, deaths, recovered, active cases
df['cum_total'] = df.groupby(by=['county'])['new_total'].transform(lambda x: x.cumsum())
df['cum_deaths'] = df.groupby(by=['county'])['new_deaths'].transform(lambda x: x.cumsum())
df['cum_recovered'] = df.groupby(by=['county'])['new_recovered'].transform(lambda x: x.cumsum())
#df.rename(columns = {'cases':'New_Total','deaths':'New_Deaths','recovered':'New_Recovered'},inplace=True)
df['cum_active']=df['cum_total']-df['cum_deaths']-df['cum_recovered']
df['new_active']=df['new_total']-df['new_deaths']-df['new_recovered']
pd.set_option('display.max_rows', None)
df.head()
```

```python
#save the county-wise case file in a csv format
df.to_csv('ger_case_corr.csv')
```
